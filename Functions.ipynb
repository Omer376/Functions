###**Q6) Explain the concept of generators in Python and how they are defined.**
###Ans: Unlike iterators, Generators uses *yield* keyword instead or return.Unlike normal functions, generators gives out result one by one instead of one go. It helps in memory management,computer computation and also helps saves time.They are defined just like a normal function using *def* keyword.

```python
def square(n):
  for i in range(n):
    yield i**2
a=square(5)
next(a)
next(a)
```

###**Q7)  What are the advantages of using generators over regular functions?**
###Ans: Normal functions perform computation first and then return the result in one go.This results in a lot of execution time and a lot of computation.Generators overcome this drawback by giving result one by one which means lesser memory usage and less execution time.


```python
#general function
def sq(n):
  l=[]
  for i in range(n):
    l.append(i**2)
  return l
sq(5)
```

```python
#generators
def sq(n):
  for i in range(n):
    yield i**2
a=sq(5)
next(a)
```

```python
"""Q1) Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list."""
def is_even(n):
  l=[]
  for i in n:
    if i%2==0:
      l.append(i)
  return sum(l)
is_even([1,2,3,4,45,5,6,7,8])
#lambda
is_even_lambda=lambda n:sum([i for i in n if i%2==0])
is_even_lambda([1,2,3,4,45,5,6,7,8])
```
